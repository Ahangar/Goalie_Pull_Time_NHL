```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import os
import re
import matplotlib.pyplot as plt
```

```python
def get_game_data(soup):
    t=[]
    away_line=[]
    home_line=[]
    event=[]
    description=[]
    n_row=[]
    period=[]

    #reading the html tables
    for row in soup.find_all('tr', {'class': ['evenColor','oddColor']}):
        
        #reading the columns from the html file        
        n_row.append(row.find_all('td')[0].text)
        period.append(row.find_all('td')[1].text)
        t.append(row.find_all('td')[3].text)
        event.append(row.find_all('td')[4].text)
        description.append(row.find_all('td')[5].text)
        
        #reading the team lines at each row
        home_players=''  #for empty cells
        away_players=''
        is_away=1        #the first team line column is the away team
        for table_row in row.find_all('table'):              
            if (len(table_row.find_all('table'))>1):
                players=' '.join(re.findall('[a-zA-Z]+', table_row.text.strip()))
                if is_away:
                    away_players=players
                    is_away=0
                else:
                    home_players=players
                        
        away_line.append(away_players)
        home_line.append(home_players)
        
    #creating the dataframe containing all of the data                                        
    df=pd.DataFrame(list(zip(n_row,t,period,event,away_line,home_line,description)),
                 columns =['n_row','time','period','event','away_line','home_line','description'])
    
    #convert time to minute and second format
    df['minute']=df['time'].apply(lambda x: x.split(':')[1][2:])
    df['second']=df['time'].apply(lambda x: x.split(':')[2])
    df=df.drop(columns='time')

    #read the team names and add it to the datafram
    team_names = []
    for row_data in soup.find_all('td', {'class': 'heading + bborder'}):
        if 'on ice' in row_data.text.lower():
            team_names.append(row_data.text.split()[0])
        if len(team_names) == 2:
             break
    if len(team_names) != 2:
        team_names = ['', '']
        
    df['away_team']=team_names[0]
    df['home_team']=team_names[1]
    
    return(df)
```

```python
season='20132014'

for file in os.listdir('./data/raw/'+season+'/'):    
    url='./data/raw/'+season+'/'+file
    with open(url) as fp:
        soup = BeautifulSoup(fp.read(), 'lxml')
    game_data = get_game_data(soup)
    game_data.to_csv('./data/game/'+season+'_'+file+'.csv',index=False)
        
    print(file)
```
